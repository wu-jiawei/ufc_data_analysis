## 列定义
* R_和B_前缀表示红角和蓝角选手对应的数据
* hst_中缀表示是在当前比赛之前的历史平均数据(场均）
* _attempted后缀表示选手的尝试数
* _succeeded后缀表示的是选手的成功数
* fighter: 选手名字
* Format: 比赛形式，包括回数和每回时间
* Referee: 场上裁判名字
* date: 比赛日期
* location: 比赛地点
* Fight_type: 比赛类型，包括比赛量级等信息
* Winner: 胜者名字
* Title: 表明比赛是否为头衔争夺赛
* weight_class: 比赛量级
* SIG_STR.: 有效打击数
* TOTAL_STR.: 全部打击数
* TD: takedowns, 摔倒对方数
* HEAD: 对对方的头部打击数
* BODY: 对对方的身体打击数
* LEG: 对对方的腿部打击数
* DISTANCE: 保持与对方的距离数
* CLINCH: 在箍颈状态下对对方的打击数
* GROUND: 在地面对对方的打击数
* KD: kickdown, 击倒对方数
* SIG_STR_pct: 有效打击成功率
* TD_pct: 摔倒成功率
* SUB_ATT: 降服尝试数
* PASS: 不清楚
* REV: 不清楚
* Stance: 站姿
* Height(cm): 身高
* Weight(kg): 体重
* Reach(cm): 臂展
* age: 年龄
* experience: 本次比赛前参加UFC次数
* title_experience: 本次比赛前参加UFC头衔争夺赛的次数
* KO: KO率
* sub: 降服率
* winning: 历史胜率
* title_winning: 头衔赛的历史胜率

### 单变量影响
来看看哪些特征对胜率有影响，先是随着参加次数的增加，胜率的变化

```python
ax = sns.boxplot(y=ptc_counts)
ax.set_title('UFC participating times')
```

只有极少数运动员参加了超过15次UFC比赛，而大多数运动员不会参加超过10次比赛

```python
tmp1 = data[['R_experience', 'R_winning']].groupby('R_experience').mean()
tmp2 = data[['B_experience', 'B_winning']].groupby('B_experience').mean()

sns.set(style='whitegrid')
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(tmp1.index, tmp1['R_winning'], 'r-', label='red corner')
ax.plot(tmp2.index, tmp2['B_winning'], 'b-', label='blue corner')
ax.set_xlabel('Participating times')
ax.set_ylabel('Winning rate')
ax.legend()
```

可以看到在前10次参加比赛时随着比赛次数的增加运动员的胜率快速提高，这可能出于以下两个原因：
* 幸存者偏差，只有前几次比赛成绩好的运动员才有资格继续比赛
* 运动员最开始的几场比赛都不太重要，因此多遇到实力，经验不太够的菜鸟

在第10到第20次参加比赛时运动员的胜率保持平稳，这可能是因为:
* 这时候因为比赛次数已经很多了，单次比赛的胜负对平均胜率的影响不大
* 这时候运动员参加的都是比较重要的比赛，组织方会为其精心挑选对手

因为很少有运动员会参加20次以上的比赛，因此此后的胜率波动较大，但似乎胜率在下降， 这可能是因为：
* 在赢了很多场比赛之后，运动员打更大重量级比赛，导致成绩下降
* 运动员年纪大了之后，竞技状态下滑
* 随着mma这项运动的逐渐完善，技术越来越趋同，运动员之间的水平差距变小

然而不论怎样，我们都可以看到，打过很多场比赛的明星运动员的胜率是要高出只打过几场比赛的菜鸟运动员很多的

```python
columns = ['SIG_STR.', 'TOTAL_STR.', 'TD', 'HEAD', 'BODY', 'LEG', 'DISTANCE', 'CLINCH', 'GROUND']
columns = ['R_hst_'+column+'_succeeded' for column in columns]
columns += ["R_hst_KD", 'R_hst_SIG_STR_pct', 'R_hst_TD_pct','R_hst_SUB_ATT','R_hst_PASS','R_hst_REV',
            'R_KO', 'R_sub']
columns += ['R_winning']
R_corr = data[columns].corr()
```

```python
fig, ax = plt.subplots(figsize=(14, 14))
ax = sns.heatmap(data=R_corr, cmap='Reds', annot=True)
ax.set_title('About red corner')
```

观察上面的pearson相关系数热力图，可以发现下面一些事实:

* 从最右边R_winning 一列可以看到，所有这些特征与历史胜率都有正的线性关系，除了REV有接近于零的之外。
* KO和降服能力强的运动员的成绩好是理所当然的，我们之后再看看这两种终结比赛的方式。
* 有效打击数是决定比赛结果的重要特征。
* 对头部的打击数重要过多躯干的打击数，而对腿部的打击数及在箍颈状态下的打击数对胜利不是太重要，这两种都是消耗，干扰对手的方式，一般对比赛结果起不到决定作用。
* 击倒对手和地面锤击都都是获胜的重要手段。
* PASS 特征也与历史胜率高度正相关，或许这一特征是表明选手的逃脱率.
* 注意到热力图的左上角区域颜色较深，这些特征都是关于打击成功率的，说明各种打击是高度相关的，运动员攻击时都是联合攻击的.

红蓝角选手的年龄差分布是以0为中心的非常对称的分布，来看年龄差距对比赛是否有影响。

```python
R_tmp = data['Winner'][data['R_age']<=data['B_age']-5]==data['R_fighter'][data['R_age']<=data['B_age']-5]
R_tmp = R_tmp.sum() / (data['R_age']<=data['B_age']-5).sum()
B_tmp = data['Winner'][data['B_age']<=data['R_age']-5]==data['B_fighter'][data['B_age']<=data['R_age']-5]
B_tmp = B_tmp.sum() / (data['B_age']<=data['R_age']-5).sum()
print('红角选手在比蓝角选手至少年轻5岁时的获胜比例：', R_tmp)
print('红角选手的平均获胜比例：', (data['R_fighter']==data['Winner']).sum()/data.shape[0])
print('\n蓝角选手在比红角选手至少年轻5岁时的获胜比例：', B_tmp)
print('蓝角选手的平均获胜比例：', (data['B_fighter']==data['Winner']).sum()/data.shape[0])
```

可以看到不论红角蓝角选手当他们比对手年轻至少5岁时获胜的比例比起平均获胜比例大大提高。这说明更年轻确实更占优。

看起来所有量级的线条搅在了一起，或是因为量级划分太细的缘故，为此考虑构造新的量级划分

```python
def get_new_weight_class(row):
    """
    row: a row of the data DataFrame
    returns; new weight class of the fight
    """
    weight = row['weight_class']
    if weight in ['Flyweight', 'Bantamweight', 'Featherweight']:
        return 'Light'
    if weight in ['Lightweight', 'Welterweight', 'Middleweight']:
        return 'Middle'
    else:
        return 'Heavy'

data['new_weight_class'] = data.apply(get_new_weight_class, axis=1)
```
